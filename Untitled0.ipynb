```
import tensorflow as tf #importamos tensor flow
import numpy as np  #ayuda a iterpretar nuestors datos como listas de manera facil
from tensorflow import keras #marco para definir una red neuronal

#crear una red de una capa y una neurona --> solo tiene una entrada de un valor
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
#perdida -> mide las respuestas adivinadas contra las respuestas correctas conocidas y mide que tan bien o que tan mal lo hizo
#optmizadora -> usa esta funcion para hacer otra suposicion, segun el resultado de la funcion de perdida, intntara minimizar la perdida
model.compile(optimizer='sgd', loss='mean_squared_error')
#datos con la que entrenaremos a la red
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

#model.fit -> entrenamineto de la red entre la relacion de X y Y
#aqui es adivinar, que tan bueno o malo es(perdida usar el optimizador para hacer otra supocision por una determinada cantidad de veces
model.fit(xs, ys, epochs=500)                                           
#ahora tenemos un modelo que ha sido entrenado para aprender la relacion entre X y Y
#el resultado no es exacto debido a que las redes no son exactas trabajn mas bien con probabilidades
print(model.predict([10,0]))
```
